```python
!pip install matplotlib pandas

import sys
!{sys.executable} -m pip install pandas matplotlib



import matplotlib.pyplot as plt
import pandas as pd

# Load datasets
models_df = pd.read_csv("all_ai_models.csv")
ai_use_cases_df = pd.read_csv("2023 Consolidated AI Use Case Inventory (PUBLIC).csv")

# Fix Year Column for AI Models
models_df['Year'] = pd.to_datetime(models_df['Publication date'], errors='coerce').dt.year
models_df = models_df.dropna(subset=['Year'])
models_df['Year'] = models_df['Year'].astype(int)
models_df = models_df[models_df['Year'] >= 2000]
ai_models = models_df.groupby('Year').size()
ai_models = ai_models.rolling(window=3, min_periods=1).mean()

# Fix Proxy Adoption Rate from Development_Stage
adoption_counts = ai_use_cases_df['Development_Stage'].value_counts()
adoption_counts = adoption_counts.rolling(window=3, min_periods=1).mean()
adoption_years = list(range(2010, 2010 + len(adoption_counts)))

# Create main figure
plt.figure(figsize=(12, 7))
plt.plot(ai_models.index, ai_models.values, marker='o', linestyle='-', label='AI Model Growth (Smoothed)', linewidth=2, color='#1f77b4')
plt.plot(adoption_years, adoption_counts.values, marker='s', linestyle='--', label='AI Adoption Rate (Smoothed)', linewidth=2, color='#ff7f0e')

# Add a secondary panel inset with bar chart of development stage counts
inset_axes = plt.axes([0.6, 0.5, 0.28, 0.35])
development_stage_counts = ai_use_cases_df['Development_Stage'].value_counts().sort_values()
inset_axes.barh(development_stage_counts.index, development_stage_counts.values, color='#2ca02c')
inset_axes.set_title('Use Case Development Stages', fontsize=8)
inset_axes.tick_params(labelsize=6)

# Labels and title
plt.xlabel('Year', fontsize=12)
plt.ylabel('Count / Adoption Rate (%)', fontsize=12)
plt.title('AI Model Growth vs AI Adoption Rate (2000-2025)', fontsize=14)
plt.legend(loc='upper left', frameon=False)
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Save figure
plt.savefig("ai_adoption_trends_final.png", dpi=300, bbox_inches='tight')
plt.show()

print("Figure saved as ai_adoption_trends_final.png")

```
